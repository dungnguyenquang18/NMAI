````python
import re
import google.generativeai as genai
import time
import os
from dotenv import load_dotenv

load_dotenv()


class LLM():
    def __init__(self, api_key=None):
        # Thay thế bằng API Key của bạn
        self.API_KEY = api_key
        # Cấu hình API Key
        genai.configure(api_key=self.API_KEY)
        # Khởi tạo mô hình Gemini Pro
        self.model = genai.GenerativeModel("gemini-2.0-flash")

    def answer(self, query):
        result = self.model.generate_content(query).text
        return result

    def extract_entities_and_relationships(self, text, state):

        prompt1 = (
            f"Bạn là một chuyên gia trong lĩnh vực coding. "
            f"Vui lòng triển khai lại code sau từ boostrap thành html thuần để ngắn gọn lại:\n"
            f"{text}"
        )
        prompt2 = (
            f"Bạn là một chuyên gia trong lĩnh vực coding. "
            f"Vui lòng viết resume ở định dạng html từ các thông tin sau:\n"
            f"{text}"
        )
        prompt3 = (
            f"Bạn là một AI assitant."
            f"Vui lòng thêm vào thông tin này các thông tin liên hệ bất kỳ bao gồm họ tên, số điện thoại và email sau đó chuyển những thông tin sau về định dạng json:\n"
            f"{text}"
        )
        for _ in range(5):  # Gửi 5 request
            try:
                response = ""
                if state == 1:
                    response = self.answer(prompt1)
                else:
                    response = self.answer(prompt3)
                return self.extract_code_block(response)
            except Exception as e:
                print(f"Lỗi: {e}")
                time.sleep(20)  # Chờ 10 giây trước khi thử lại

        return 'lỗi api'
    def extract_code_block(self, text):
        match = re.search(r"```(?:\w*\n)?(.*?)```", text, re.DOTALL)
        if match:
            return match.group(1).strip()
        return None
    
````
